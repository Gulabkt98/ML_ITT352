```python
import torch

# Initialize values
x = torch.tensor(2.0)
y_true = torch.tensor(10.0)

# Parameters
W = torch.tensor(1.0, requires_grad=False)
b = torch.tensor(0.0, requires_grad=False)

lr = 0.01

for i in range(50):
    # Forward
    y_pred = W * x + b
    loss = 0.5 * (y_pred - y_true)**2

    # Manual gradients
    dL_dy = (y_pred - y_true)
    dL_dW = dL_dy * x
    dL_db = dL_dy * 1

    # Update
    W = W - lr * dL_dW
    b = b - lr * dL_db

    if i % 10 == 0:
        print(f"Iteration {i}, Loss = {loss.item()}")
```
